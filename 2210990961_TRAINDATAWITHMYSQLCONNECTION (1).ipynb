# HOW MANY PEOPLE SURVIVED AND HOW MANY DIED PLOT ON GRAPH

```python
import matplotlib.pyplot as plt
import numpy as np
import mysql.connector as sql
import pandas as pd

db_connection = sql.connect(host='localhost', database='train', user='root', password='goodboy@123#')
query = "select * from train"
df_train = pd.read_sql(query, con = db_connection)


survive = len(df_train[df_train["Survived"]==1])
nsurvive = len(df_train[df_train["Survived"]==0])
x = ['survived','not survived']
y = np.array([survive, nsurvive])

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(y, labels=x,autopct='%1.2f%%')
plt.show()
```

# HOW MANY PASSENGERS WERE IN FIRST CLASS,SECOND CLASS AND THIRD CLASS PLOT THOSE FIGURES ON GRAPH

```python
import matplotlib.pyplot as plt
import numpy as np
import mysql.connector as sql
import pandas as pd

db_connection = sql.connect(host='localhost', database='train', user='root', password='goodboy@123#')
query = "select * from train"
df_train = pd.read_sql(query, con = db_connection)


class_1 = len(df_train[df_train["Pclass"]==1])
class_2 = len(df_train[df_train["Pclass"]==2])
class_3 = len(df_train[df_train["Pclass"]==3])
x = ['first class','second class','third class']
y = np.array([class_1,class_2,class_3])

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(y, labels=x,autopct='%1.2f%%')
plt.show()
```

# BIVARIANT ANALSIS

```python
#when we represent two variables on a graph, that is knwown as bivaraite analysis
data.columns
plt.figure(figsize=(8, 6))
plt.scatter(data['PassengerId'], data['Age'], color='blue', alpha=0.7)
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('Scatter Plot of Variable 1 vs Variable 2')
plt.show()
```

# HOW HAS BETTER CANCE OF SURVIVAL MALE OR FEMALE

```python
survival_pivot = data.pivot_table(index='Sex', columns='Survived', aggfunc='size', fill_value=0)

# Plot the survival rates
survival_pivot.plot(kind='bar', stacked=True, figsize=(8, 6), color=['red', 'green'])
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Survival Count by Sex')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.xticks(rotation=0)
plt.show()
```
