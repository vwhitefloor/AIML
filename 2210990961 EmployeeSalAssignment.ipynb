# FIND MINIMUM,MAXIMUM AND AVERGE BASEPAY

```python
data['BasePay'] = pd.to_numeric(data['BasePay'], errors='coerce')
data['BasePay'].min()
data['BasePay'].max()
data['BasePay'].mean()
```

# REPLACE 'NOT PROVIDED' IN  EMPLOYEENAME COLUMN TO NaN

```python
import numpy as np

data['EmployeeName'] = data['EmployeeName'].replace('NOT PROVIDED', np.nan)
```

# drop the rows having more than 5 missing values

```python
data.dropna(thresh=5)
```

# find the job title of ALBERT PARDINI

```python
albert_pardini_job_title = data[data['EmployeeName'] == 'ALBERT PARDINI']['JobTitle'].iloc[0]
print("Job Title of ALBERT PARDINI:", albert_pardini_job_title)
```

# HOW MUCH ALBERT PARDINI MAKE (INCULDE DENEFITS)?

```python
relevant_columns = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']
data[relevant_columns] = data[relevant_columns].apply(pd.to_numeric, errors='coerce')
albert_pardini_data = data[data['EmployeeName'] == 'ALBERT PARDINI']
total_compensation = albert_pardini_data[relevant_columns].sum(axis=1).iloc[0]

print("Total Compensation for ALBERT PARDINI:", total_compensation)
```

# DISPLAY NAME OF THE PERSON HAVINNG THE HIGHEST BASEPAY

```python
data['Basepay'] = pd.to_numeric(data['BasePay'], errors='coerce')
max_basepay_index = data['BasePay'].idxmax()
person_highest_basepay = data.loc[max_basepay_index, 'EmployeeName']

print("Person with the highest base pay:", person_highest_basepay)
```

# FIND AVERGE BASEPAY OF ALL EMPLOYEE PER YEAR

```python
data['Year'] = pd.to_datetime(data['Year'], errors='coerce').dt.year
average_basepay_per_year = data.groupby('Year')['BasePay'].mean()
print("Average Base Pay per Year:")
print(average_basepay_per_year)
```

# FIND AVERGE BASEPAY OF ALL EMPLOYEE PER JOB TITLE

```python
data.groupby('JobTitle')['BasePay'].mean()
```
